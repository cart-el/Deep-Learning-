```python
# Datatype of the column appears to be a dictionary but on
# trying to unpack it, an empty column was given as output /n
# indicating the likelihood of the entries being stored as a 
# string representation of a dictionary, hence the need to cast 
# the string representations as a dictionaries.

# Convert string representations to dictionaries
dataset['sentiment'] = dataset['sentiment'].apply(ast.literal_eval)
```

```python
# Unpack the dictionary content (key-value pairs) into new columns
dict_unpacked = pd.json_normalize(dataset['sentiment'])
dict_unpacked
```

```python
# Concatenate the unpacked columns with the original DataFrame
crypto_df = pd.concat([dataset, dict_unpacked], axis=1)

# Drop the original dictionary column
crypto_df.drop(columns=['sentiment'], inplace=True)

crypto_df
```
